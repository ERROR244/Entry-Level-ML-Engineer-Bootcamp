```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Week00/solution/ex07/data.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data = data.dropna(subset=['price']).reset_index(drop=True)

print(data['price'].isna().sum())
```

```python
data['area'] = data['area'].fillna(data['area'].mode()[0])
data['stories'] = data['stories'].fillna(data['stories'].mode()[0])
data['airconditioning'] = data['airconditioning'].fillna(data['airconditioning'].mode()[0])
data['parking'] = data['parking'].fillna(data['parking'].mode()[0])
data['prefarea'] = data['prefarea'].fillna(data['prefarea'].mode()[0])
data['furnishingstatus'] = data['furnishingstatus'].fillna(data['furnishingstatus'].mode()[0])
```

```python
data.isna().sum()
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_data = encoder.fit_transform(data[['furnishingstatus']])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['furnishingstatus']))
data.drop('furnishingstatus', axis=1, inplace=True)
pd.concat([data, encoded_df], axis=1)

label_encoder = preprocessing.LabelEncoder()
data['mainroad'] = label_encoder.fit_transform(data['mainroad'])
data['guestroom'] = label_encoder.fit_transform(data['guestroom'])
data['basement'] = label_encoder.fit_transform(data['basement'])
data['hotwaterheating'] = label_encoder.fit_transform(data['hotwaterheating'])
data['airconditioning'] = label_encoder.fit_transform(data['airconditioning'])
data['prefarea'] = label_encoder.fit_transform(data['prefarea'])
data.dtypes
```
